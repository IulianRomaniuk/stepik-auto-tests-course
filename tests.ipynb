```python
from selenium import webdriver
import time

try: 
    link = "http://suninjuly.github.io/registration2.html"
    browser = webdriver.Chrome()
    browser.get(link)

    # Ваш код, который заполняет обязательные поля
    input1 = browser.find_element_by_xpath('//input [@placeholder = "Input your first name"]')
    input1.send_keys("Ok")
    input2 = browser.find_element_by_xpath('//input [@placeholder = "Input your last name"]')
    input2.send_keys("Ok")
    input3 = browser.find_element_by_xpath('//input [@placeholder = "Input your email"]')
    input3.send_keys("Ok")

    # Отправляем заполненную форму
    button = browser.find_element_by_css_selector("button.btn")
    button.click()

    # Проверяем, что смогли зарегистрироваться
    # ждем загрузки страницы
    time.sleep(5)

    # находим элемент, содержащий текст
    welcome_text_elt = browser.find_element_by_tag_name("h1")
    # записываем в переменную welcome_text текст из элемента welcome_text_elt
    welcome_text = welcome_text_elt.text

    # с помощью assert проверяем, что ожидаемый текст совпадает с текстом на странице сайта
    assert "Congratulations! You have successfully registered!" == welcome_text

finally:
    # ожидание чтобы визуально оценить результаты прохождения скрипта
    time.sleep(10)
    # закрываем браузер после всех манипуляций
    browser.quit()
```

```python
from selenium import webdriver
import math
import time

link = "http://suninjuly.github.io/get_attribute.html"

try:
    browser = webdriver.Chrome()
    browser.get(link)
    
    box = browser.find_element_by_tag_name("img")
    x_element = box.get_attribute("valuex")
    def calc(x):
        return str(math.log(abs(12*math.sin(int(x)))))
    x = x_element
    y = calc(x)
    
    input1 = browser.find_element_by_css_selector("input[id = 'answer']")
    input1.send_keys(y)
    
    checkbox = browser.find_element_by_css_selector("input[type = 'checkbox']")
    checkbox.click()
    
    radio = browser.find_element_by_css_selector("input[id = 'robotsRule']")
    radio.click()
    
    submit = browser.find_element_by_css_selector("button[type = 'submit']")
    submit.click()
    
finally:
    time.sleep(10)
    browser.quit()
```

```python
from selenium import webdriver
import time
import math

link = "http://suninjuly.github.io/selects2.html"

try:
    browser = webdriver.Chrome()
    browser.get(link)
    
    x = browser.find_element_by_css_selector("span[id = 'num1']")
    y = browser.find_element_by_css_selector("span[id = 'num2']")
    
    first = int(x.text) 
    second = int(y.text) 
    
    def summ(a,b):
        return str(first + second)
    ans = summ(first,second) 
    
    #print (ans)
    
    from selenium.webdriver.support.ui import Select
    select = Select(browser.find_element_by_tag_name("select"))
    
    browser.find_element_by_css_selector("#dropdown").click() #click on dropdown
    select.select_by_value(ans)
    browser.find_element_by_css_selector("button[type = 'submit'").click()
    
finally:
    time.sleep(20)
    browser.quit()
   
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver

browser = webdriver.Chrome()

browser.get("http://suninjuly.github.io/wait2.html")

# говорим Selenium проверять в течение 5 секунд, пока кнопка не станет кликабельной
button = WebDriverWait(browser, 5).until(
        EC.element_to_be_clickable((By.ID, "verify"))
    )
button.click()
message = browser.find_element_by_id("verify_message")

assert "successful" in message.text
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
import time

link = 'http://suninjuly.github.io/explicit_wait2.html'

try:
    driver = webdriver.Chrome()
    driver.get(link)
    WebDriverWait(driver, 15).until(EC.text_to_be_present_in_element((By.ID,"price"), "$100"))
    driver.find_element_by_css_selector("button[id = 'book']").click()
    driver.execute_script("window.scrollTo(0, 500)")
    num = driver.find_element_by_css_selector("span[id ='input_value']")
    x =int(num.text)
    
    def calc(respons):
        return str(math.log(abs(12*math.sin(int(x)))))
    respons = calc(x)
    
    input1 = driver.find_element_by_tag_name('input')
    input1.send_keys(respons) 
    
    driver.find_element_by_css_selector("button[id = 'solve']").click()
   
finally:
    time.sleep(8)
    driver.quit()
```
